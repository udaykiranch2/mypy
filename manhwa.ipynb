```python
import pandas
from bs4 import BeautifulSoup
import requests
import lxml
```

```python
# base_url = "https://asuratoon.com"
cells = []
pages  =10
for i in range(1,pages+1):
    print(f"obtaing pages {i}")
    url = f"https://asuratoon.com/manga/?page={i}&status=&type=&order=popular"
    response = requests.get(url).content
    
    soup = BeautifulSoup(response,'lxml')
    content = soup.find_all("div",class_="bigor")
    cells.extend(content)

    
    
```

```python
for cell in cells :
    print(cell.prettify())
```

```python
names = []
ratings = []
chapters = []
```

```python
for cell in cells:

    name = cell.find("div",class_="tt").get_text().strip()
    chapter = cell.find("div",class_="epxs").get_text().strip()
    rating = cell.find("div",class_="numscore").get_text().strip()
    names.append(name)
    ratings.append(rating)
    chapters.append(chapter)
    
```

```python
print(len(names), len(ratings), len(chapters))
```

```python
mydict = {"Name":names,
          "Chapters":chapters,
          "Rating":ratings}
```

```python
df = pandas.DataFrame(mydict)
```
